```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
BGR = cv2.imread("./RightLeft/2/two192.jpg")              #

b,g,r = cv2.split(BGR)	#

RGB= cv2.merge([r,g,b])  #RGB

plt.imshow(RGB) #RGB
plt.show()
```

```python
BGR = cv2.imread("./BLACK/2/two192.jpg")              #

b,g,r = cv2.split(BGR)	#

RGB= cv2.merge([r,g,b])  #RGB

plt.imshow(RGB) #RGB
plt.show()
```

```python
hsi= cv2.cvtColor(RGB, cv2.COLOR_RGB2HSV)  #HSI

h, s, i = cv2.split(hsi)        #HSI
```

```python
saturation_threshold = 100  # 

# Saturation Slicing
saturation_mask = s < saturation_threshold
saturation_mask = saturation_mask.astype(int)

saturation_mask = saturation_mask -1


#  mask 
segmented = cv2.bitwise_and(BGR, BGR, mask=saturation_mask.astype(np.uint8))

# 
plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(BGR, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(1, 4, 2)
plt.imshow(s, cmap='gray')
plt.title('Saturation Channel')

plt.subplot(1, 4, 3)
plt.imshow(saturation_mask)
plt.title('Saturation Mask')

plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(segmented, cv2.COLOR_BGR2RGB))
plt.title('Segmented Image')

plt.show()
```

```python
# 
input_folder = f"./zerotofour/3"
output_folder = f"./black0-4/3"

# 
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 
for filename in os.listdir(input_folder):
    if filename =="desktop.ini":
        continue
    input_path = os.path.join(input_folder, filename)
    output_path = os.path.join(output_folder, filename)
    
    print(input_path)
    # cv2
    img = cv2.imread(input_path)
    b,g,r = cv2.split(img)	#
    RGB= cv2.merge([r,g,b])  #RGB
    hsi= cv2.cvtColor(img, cv2.COLOR_RGB2HSV)  #HSI
    h, s, i = cv2.split(hsi)        #HSI
    
    saturation_threshold =70 # 

    # Saturation Slicing
    saturation_mask = s < saturation_threshold
    saturation_mask = saturation_mask.astype(int)

    saturation_mask =1- saturation_mask 

    # 
    cv2.imwrite(output_path, saturation_mask* 255)
```
