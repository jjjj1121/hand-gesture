```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
BGR = cv2.imread("./RightLeft/2/two192.jpg")              #將圖片讀入

b,g,r = cv2.split(BGR)	#將圖片拆成單通道

RGB= cv2.merge([r,g,b])  #在將圖片按照RGB順序組裝回去

plt.imshow(RGB) #顯示RGB原圖
plt.show()
```

```python
BGR = cv2.imread("./BLACK/2/two192.jpg")              #將圖片讀入

b,g,r = cv2.split(BGR)	#將圖片拆成單通道

RGB= cv2.merge([r,g,b])  #在將圖片按照RGB順序組裝回去

plt.imshow(RGB) #顯示RGB原圖
plt.show()
```

```python
hsi= cv2.cvtColor(RGB, cv2.COLOR_RGB2HSV)  #將圖片轉成HSI格式

h, s, i = cv2.split(hsi)        #將圖片拆成分別為HSI的單通道圖片
```

```python
saturation_threshold = 100  # 調整需要的閾值

# Saturation Slicing
saturation_mask = s < saturation_threshold
saturation_mask = saturation_mask.astype(int)

saturation_mask = saturation_mask -1


# 最終 mask 應用在原圖上
segmented = cv2.bitwise_and(BGR, BGR, mask=saturation_mask.astype(np.uint8))

# 顯示圖片
plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(BGR, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(1, 4, 2)
plt.imshow(s, cmap='gray')
plt.title('Saturation Channel')

plt.subplot(1, 4, 3)
plt.imshow(saturation_mask)
plt.title('Saturation Mask')

plt.subplot(1, 4, 4)
plt.imshow(cv2.cvtColor(segmented, cv2.COLOR_BGR2RGB))
plt.title('Segmented Image')

plt.show()
```

```python
# 指定輸入和輸出資料夾的路徑
input_folder = f"./zerotofour/3"
output_folder = f"./black0-4/3"

# 確保輸出資料夾存在
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 遍歷輸入資料夾中的所有圖片
for filename in os.listdir(input_folder):
    if filename =="desktop.ini":
        continue
    input_path = os.path.join(input_folder, filename)
    output_path = os.path.join(output_folder, filename)
    
    print(input_path)
    # 使用cv2讀取圖片
    img = cv2.imread(input_path)
    b,g,r = cv2.split(img)	#將圖片拆成單通道
    RGB= cv2.merge([r,g,b])  #在將圖片按照RGB順序組裝回去
    hsi= cv2.cvtColor(img, cv2.COLOR_RGB2HSV)  #將圖片轉成HSI格式
    h, s, i = cv2.split(hsi)        #將圖片拆成分別為HSI的單通道圖片
    
    saturation_threshold =70 # 調整需要的閾值

    # Saturation Slicing
    saturation_mask = s < saturation_threshold
    saturation_mask = saturation_mask.astype(int)

    saturation_mask =1- saturation_mask 

    # 寫入灰度圖片
    cv2.imwrite(output_path, saturation_mask* 255)
```
